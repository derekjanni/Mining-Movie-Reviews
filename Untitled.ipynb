```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = pd.Series(np.random.randn(200), name='y')
```

```python
x = np.random.rand(200,40)
```

```python
%matplotlib inline
def rsquared(size):
    y = pd.Series(np.random.randn(200), name='y')
    x = np.random.rand(200, size)
    df = pd.DataFrame(data=x)
    model = sm.OLS(y, x) 
    results = model.fit()
    return results.rsquared, results.pvalues, results.fvalue
rsquared(100)

rs = [rsquared(i)[0] for i in [i for i in range(20, 200, 20)]]
ps = [np.mean(rsquared(i)[1]) for i in [i for i in range(20, 200, 20)]]
fs = [rsquared(i)[2] for i in [i for i in range(20, 200, 20)]]
xs = list(np.arange(len(rs)))
res = pd.DataFrame(index=xs, data={'R^2': rs,'P_val': ps, 'F-stat': fs})
res.plot()
```
